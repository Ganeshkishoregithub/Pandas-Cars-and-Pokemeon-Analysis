```python
#Rename columns
cars=cars.rename(columns={'Unnamed:0':'Model'})
cars
```

```python
#If we have null(blank) values in any 
cars.qsec=cars.qsec.fillna(cars.qsec.mean())
```

```python
# Drop a columns (generally we drop S.No as it dont has value)
cars=cars.drop(columns=['S.No'])
```

```python
#find corelation
co=cars[['mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']].corr()
co

#Here diagonals will have -1
#negative value indicates that they are inversely proportional
```

```python
#Stackplot
import matplotlib.pyplot as plt
%matplotlib inline
y1=cars['hp']
y2=cars['disp']
x=range(32)
plt.plot(x,y1, linewidth=2.0, color='c')
plt.stackplot(x,y1, color='red', alpha=0.7)
plt.plot(x,y2, linewidth=5.0, color='r')
plt.stackplot(x,y2, color='green', alpha=0.5)
```
